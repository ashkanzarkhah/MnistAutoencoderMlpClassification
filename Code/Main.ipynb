```python
#First we load our data in X_train, y_train, X_test, y_test
import pandas as pd

df_train = pd.read_csv("../Dataset/mnist_dataset/labeled_train_set.csv")

columns_train = df_train.columns
X_train = df_train.filter(items = columns_train[1:])
y_train = df_train.filter(items = columns_train[:1])

df_test = pd.read_csv("../Dataset/mnist_dataset/test_set.csv")

columns_test = df_test.columns
X_test = df_test.filter(items = columns_test[1:])
y_test = df_test.filter(items = columns_test[:1])
```

```python
#Now it's time to preprocess our data
#First we check if there is any missing values
print("Number of missing values in total:", df_train.isnull().sum().sum())
```

```python
#Second we check each variables interval to see if we need scaling
for column in columns_train[1:]:
    print(column, "Max:", X_train[column].max(), "Min:", X_train[column].min())
```

```python
#As we can see, they vary a lot, so I decide to do a min-max scale on it and make their intervals [0, 1]
import numpy as np

mn = 100000000; mx = 0
for column in columns_train[1:]:
    mx = max(X_train[column].max(), mx)
    mn = min(X_train[column].min(), mn)

rows = np.array(X_train) - mn
rows = rows / (mx - mn)

X_train = pd.DataFrame(rows, columns = columns_train[1:])

rows = np.array(X_test) - mn
rows = rows / (mx - mn)

X_test = pd.DataFrame(rows, columns = columns_train[1:])
```

```python
#Third we check our data's balancement
cnt = 10 * [0]
for i in range(len(y_train)):
    cnt[y_train.iloc[i][0]] += 1
    
for i in range(10):
    print("#lable", i, " :", cnt[i])
```

```python
#We can see that our data has some imbalancement but because it's not that much, I just decide to over sample lable 5 to 1750
cntt = 1750 - 1577
X_rows = np.array(X_train)
y_rows = np.array(y_train)

for i in range(len(y_train)):
    if(y_train.iloc[i][0] == 5):
        X_rows = np.append(X_rows, [X_rows[i]], axis = 0)
        y_rows = np.append(y_rows, [[5]], axis = 0)
        cntt -= 1
    if(cntt == 0):
        break
        
X_train = pd.DataFrame(X_rows, columns = columns_train[1:])
y_train = pd.DataFrame(y_rows, columns = columns_train[:1])
```

```python
#Now it's time to learn our autoencoder
import keras
from keras import layers

#Defining input shape
input_img = keras.Input(shape=(784,))

#We use 3 layers for encoding and then 3 layers for decoding
encoded = layers.Dense(392, activation='relu')(input_img)
encoded = layers.Dense(196, activation='relu')(encoded)
encoded = layers.Dense(98, activation='relu')(encoded)
encoded = layers.Dense(49, activation='relu')(encoded)

decoded = layers.Dense(98, activation='relu')(encoded)
decoded = layers.Dense(196, activation='relu')(decoded)
decoded = layers.Dense(392, activation='relu')(decoded)
decoded = layers.Dense(784, activation='sigmoid')(decoded)

#Defining model's input and output shape
autoencoder = keras.Model(input_img, decoded)

#Fixing optimizer and loss function
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

#Fitting our encoder base on Train data and loss value for both of our train and test data
autoencoder.fit(X_train, X_train,
                epochs=100,
                batch_size=256,
                shuffle=True,
                validation_data=(X_test, X_test))
```

```python
#Defining encoder and decoder base on learned encoding layers
encoder = keras.Model(input_img, encoded)
decoder = keras.Model(encoded, decoded)

#Encoding train and test images
encoded_train_imgs = encoder.predict(X_train)
encoded_test_imgs = encoder.predict(X_test)

#Decoding train and test images
decoded_train_imgs = decoder.predict(encoded_train_imgs)
decoded_test_imgs = decoder.predict(encoded_test_imgs)
```

```python
#Ploting First 20 test images and decoded one of them
import matplotlib.pyplot as plt

n = 20
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(np.array(X_test.iloc[i]).reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display reconstruction
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_test_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

```python
#Perfoming 5-fold cross-validation to find best parameters for our mlp model
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_validate

#Model paramters
Hidden_layer_sizes = [(100,), (100, 25)]
Learning_rates = ['constant', 'invscaling', 'adaptive']
Activations = ['identity', 'logistic', 'tanh', 'relu']
#Batch_sizes = [200, 300]
#Max_iters = [200, 250]

best_Hidden_layer_size = 0; best_Learning_rate = 0; best_Activation = 0; best_score = 0
#best_Batch_size = 0; best_max_iter = 0

#Learning for each set of parameters and peacking best one
for Hidden_layer_size in Hidden_layer_sizes:
    for Learning_rate in Learning_rates:
        for Activation in Activations:
            #for Batch_size in Batch_sizes:
                #for Max_iter in Max_iters:
            
            my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                                  ('model', MLPClassifier(hidden_layer_sizes=Hidden_layer_size,
                                                         learning_rate=Learning_rate,
                                                         activation = Activation))
                                                         #batch_size = Batch_size,
                                                         #max_iter = Max_iter))
                                 ])
            
            
            scores = cross_validate(my_pipeline, encoded_train_imgs, 
                    np.array([x[0] for x in np.array(y_train)]),
                    cv=5,
                    scoring=['accuracy', 'precision_macro', 'recall_macro', 'f1_macro'],
                    return_train_score=True, return_estimator=True)
            
            
            accurancy_scores = scores['test_accuracy']; precision_scores = scores['test_precision_macro']
            recall_scores = scores['test_recall_macro']; f1_scores = scores['test_f1_macro']
            
            print("Hidden layer size: ", Hidden_layer_size); print("Learning rate: ", Learning_rate)
            print("Activation: ", Activation)
            #print("Batch size: ", Batch_size); print("Max iter: ", Max_iter)
            
            print("accurancy_scores:\n", accurancy_scores); print("precision_scores:\n", precision_scores)
            print("recall_scores:\n", recall_scores); print("f1_scores:\n", f1_scores)
            
            score = np.mean(accurancy_scores)
            print("Average accurancy score:", score)
            
            if(score > best_score):
                best_Hidden_layer_size = Hidden_layer_size; best_Learning_rate = Learning_rate
                best_Activation = Activation; best_score = score
                #best_Batch_size = Batch_size; best_max_iter = Max_iter


print("best score: ", best_score)                
print("best Hidden layer size: ", best_Hidden_layer_size)
print("best Learning rate: ", best_Learning_rate)
print("best Activation: ", best_Activation)
#print("best Batch size: ", best_Batch_size)
#print("best Max iter: ", best_Max_iter)
```
